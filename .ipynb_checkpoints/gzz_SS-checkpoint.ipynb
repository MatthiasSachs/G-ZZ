```julia
using Distributions, TimeIt, ProgressMeter, PyPlot, JLD
include("zz_structures.jl")
include("mbsampler.jl")
include("gzz_temp.jl")
```

#### Load data:

```julia
dat = "GZZ_data4.jld"
X = load(dat, "X")
y = load(dat, "y")
ξ_true = load(dat, "xi_true");

d, Nobs = size(X)
mb_size = Int(round(Nobs/2));
```

#### Define prior:

```julia
σ02 = 1
prior = SS_prior(d, σ02);
# prior = gaussian_prior_nh(d, σ02);
```

#### Define model:

```julia
my_ll = ll_logistic(X,y);
my_model = model(my_ll, prior)
root = find_root(my_model, rand(d));
```

#### Define minibatch sampler:

```julia
# Sub-Sampling without control variate and with weights
ϵ = 1e-2
weights = abs.(X) + ϵ
weights ./= sum(weights,2)
gs = [wumbsampler(Nobs, mb_size, weights[i,:]) for i in 1:d]
gs_list = mbsampler_list(d,gs);
```

#### Define output scheduler:

```julia
A = eye(d)

opf = projopf(A, 1000, hyperparam_size(prior))
opt = maxa_opt(10^7)
outp = outputscheduler(opf,opt)
bb = linear_bound(my_model.ll, my_model.pr, gs_list)
update_bound(bb, my_ll, prior, gs_list, zz_state(opf));
```

#### Define mbsampler + block Gibbs sampler list:

```julia
adapt_speed = false
L = 1
my_zz_sampler = zz_sampler(0, gs_list, bb, L, adapt_speed)
hyper_sampler = block_gibbs_sampler(1e1)
blocksampler = Array{msampler}(2)
blocksampler[1] = my_zz_sampler
blocksampler[2] = hyper_sampler;
```

#### Run sampler:

```julia
start = time()
ZZ_block_sample(my_model::model, outp::outputscheduler, blocksampler::Array{msampler})
print( round((time()-start)/60, 2), " mins")
```

* Number of bounces:

```julia
zz_bounces = sum(zz_state(outp.opf).n_bounces)
hyp_bounces = outp.opf.tcounter-sum(zz_state(outp.opf).n_bounces)
@printf("ZZ bounces: %i, hyper bounces: %i", zz_bounces, hyp_bounces)
```

```julia
size(outp.opf.xi_skeleton,2)/outp.opt.max_attempts
```

#### Extract samples:

```julia
discard = 0
n_samples = 10^4
xi_samples = extract_samples(outp.opf.xi_skeleton[:,discard+1:end], 
                             outp.opf.bt_skeleton[:,discard+1:end], 
                             outp.opf.bt_skeleton[end]/n_samples,
                             "linear")[:,1:end-1];
hyp_samples = extract_samples(outp.opf.hyper_skeleton[:,discard+1:end], 
                              outp.opf.bt_skeleton[:,discard+1:end], 
                              outp.opf.bt_skeleton[end]/n_samples,
                              "constant")[:,1:end-1];
```

```julia
effort_gzz = outp.opt.max_attempts*d + zz_bounces*1*mb_size + hyp_bounces*d
effort_hmc = n_samples*100*Nobs*d
print(effort_hmc/effort_gzz)
```

#### Plot ACFs:

```julia
maxlag = 1000
acfs_xi = zeros(d, maxlag)
acfs_hyp = zeros(size(hyp_samples,1), maxlag)

for i in 1:d 
    acfs_xi[i,:] = acf(xi_samples[i,:], maxlag)
end
for i in 1:size(hyp_samples,1)
    acfs_hyp[i,:] = acf(hyp_samples[i,:], maxlag)
end
```

```julia
fig = figure("pyplot_subplot_mixed", figsize=(12,3))
subplot(121)
for i in 1:d 
    plot(acfs_xi[i,:])
end
grid(true)
title("ACF's for parameters")

subplot(122)
for i in 1:size(hyp_samples,1) 
    plot(acfs_hyp[i,:])
end
grid(true)
title("ACF's for hyperparameters");
```

```julia
mean(n_samples./sum(acfs_xi,2)), mean(abs.(n_samples./sum(acfs_xi,2)))
```

```julia
maximum(n_samples./sum(acfs_xi,2)), minimum(n_samples./sum(acfs_xi,2))
```

```julia
maximum(sum(acfs_xi,2)), minimum(sum(acfs_xi,2))
```

#### Posterior coverage:

```julia
cover = zeros(d)
ci = zeros(d,2)
for i in 1:d 
    ci[i,:] = percentile(xi_samples[i,:], [2.5, 97.5])
    cover[i] = (ci[i,1]<ξ_true[i])&(ξ_true[i]<ci[i,2])
end
```

```julia
100*mean(cover)
```

```julia
fig = figure("pyplot_barplot",figsize=(12,3))
for i in 1:d 
    plot([i,i], ci[i,:], color="blue", linewidth=1)
end
plot(1:d , ξ_true, "o", markersize=4, color="red", markersize=2)
grid(true)
xlabel("Dimension")
ylabel("CI")
title("90% confidence intervals");
```

### Run for different datasets:

```julia
xi_samples_SS, ξ_true_SS, acfs_SS, cover_SS, ci_SS = [], [], [], [], [], []
dat_files = ["GZZ_data1.jld", "GZZ_data2.jld", "GZZ_data3.jld", "GZZ_data4.jld", "GZZ_data5.jld"]
hyp_λ = repeat([1.], 5)
n_samples = repeat([10^4], 5)
mb_size = [25, 50, 75, 100, 125]
max_attempts = repeat([10^7], 5)
maxlag = 1000
```

```julia
start = time()
for i in 1:length(dat_files) 
    # spike and slab prior:
    xi_samples, ξ_true, acfs, cover, ci = 
    run_shrpr(dat_files[i], mb_size[i], max_attempts[i], hyp_λ[i], n_samples[i], "SS", maxlag)
    push!(xi_samples_SS, xi_samples)
    push!(ξ_true_SS, ξ_true)
    push!(acfs_SS, acfs)
    push!(cover_SS, cover)
    push!(ci_SS, ci)
    
end
print("Total time: ", round((time()-start)/60,2), " min")
```

```julia
for i in 1:length(dat_files) 
    @printf("coverage for SS = %i percent \n", 100*mean(cover_SS[i]))
end
```
