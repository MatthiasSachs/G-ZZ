```julia
fig = figure("pyplot_barplot",figsize=(12,4))
n_samples = 10^4
maxlag = 1000
discard = 50_000
iacts_lambda = zeros(10)
iacts_mb = zeros(10)
for j in 1:2 
    for i in 1:5
        k = (j-1)*5+i
        filename = "/xtmp/GZZ_data/shrinkage_prior/gastroenterology/lambda:"*
                    string(lambda_list[k])*"-mb_size:"*string(mb_size)*".jld"
        xt_skel, bt_skel = load(filename, "xt_skeleton", "bt_skeleton")
        xi_samples = extract_samples(xt_skel[:,discard+1:end], 
                                     bt_skel[:,discard+1:end], 
                                    (bt_skel[end]-bt_skel[discard])/n_samples,
                                    "linear")[:,1:end-1];
        iacts_lambda[k] = maximum([sum(acf(xi_samples[dim,:],maxlag)) for dim in 1:size(xi_samples,1)])
        
        filename = "/xtmp/GZZ_data/shrinkage_prior/gastroenterology/lambda:"*
                    string(lambda)*"-mb_size:"*string(mb_size_list[k])*".jld"
        xt_skel, bt_skel = load(filename, "xt_skeleton", "bt_skeleton")
        xi_samples = extract_samples(xt_skel[:,discard+1:end], 
                                     bt_skel[:,discard+1:end], 
                                    (bt_skel[end]-bt_skel[discard])/n_samples,
                                    "linear")[:,1:end-1];
        iacts_mb[k] = maximum([sum(acf(xi_samples[dim,:],maxlag)) for dim in 1:size(xi_samples,1)])
        xt_skel, bt_skel, xi_samples = [], [], []
        gc()
    end
end
```

```julia
fig = figure("pyplot_barplot",figsize=(12,3))
n_samples = 10^4
maxlag = 500
discard = 50_000
iacts_lambda = zeros(10)
iacts_mb = zeros(10)
for j in 1:2 
    for i in 1:5
        k = (j-1)*5+i
        filename = "/xtmp/GZZ_data/shrinkage_prior/gastroenterology/lambda:"*
                    string(lambda_list[k])*"-mb_size:"*string(mb_size)*".jld"
        xt_skel, bt_skel = load(filename, "xt_skeleton", "bt_skeleton")
        xi_samples = extract_samples(xt_skel[:,discard+1:end], 
                                     bt_skel[:,discard+1:end], 
                                    (bt_skel[end]-bt_skel[discard])/n_samples,
                                    "linear")[:,1:end-1];
        iacts_lambda[k] = maximum([sum(acf(xi_samples[dim,:],maxlag)) for dim in 1:size(xi_samples,1)])
        subplot(2,5,k) 
        for dim in 1:size(xi_samples,1)
            plot(acf(xi_samples[dim,:],maxlag))
        end
        grid(true)
        title("lambda = "*string(round(lambda_list[k],2)), fontsize=10)
        if i == 1 ylabel("ACF") end
        if j == 2 xlabel("Lag") end
        if j == 1 xticks(alpha=0) end
        subplots_adjust(hspace=0.3, wspace=0.25)
        xt_skel, bt_skel, xi_samples = [], [], []
        gc()
    end
end
```

```julia
fig = figure("pyplot_barplot",figsize=(12,3))
n_samples = 10^4
maxlag = 500
discard = 50_000
iacts_lambda = zeros(10)
iacts_mb = zeros(10)
for j in 1:2 
    for i in 1:5
        k = (j-1)*5+i
        filename = "/xtmp/GZZ_data/shrinkage_prior/gastroenterology/lambda:"*
                    string(lambda)*"-mb_size:"*string(mb_size_list[k])*".jld"
        xt_skel, bt_skel = load(filename, "xt_skeleton", "bt_skeleton")
        xi_samples = extract_samples(xt_skel[:,discard+1:end], 
                                     bt_skel[:,discard+1:end], 
                                    (bt_skel[end]-bt_skel[discard])/n_samples,
                                    "linear")[:,1:end-1];
        iacts_lambda[k] = maximum([sum(acf(xi_samples[dim,:],maxlag)) for dim in 1:size(xi_samples,1)])
        subplot(2,5,k) 
        for dim in 1:size(xi_samples,1)
            plot(acf(xi_samples[dim,:],maxlag))
        end
        grid(true)
        title("mb_size = "*string(mb_size_list[k]), fontsize=10)
        if i == 1 ylabel("ACF") end
        if j == 2 xlabel("Lag") end
        if j == 1 xticks(alpha=0) end
        subplots_adjust(hspace=0.3, wspace=0.25)
        xt_skel, bt_skel, xi_samples = [], [], []
        gc()
    end
end
```
