```julia
coverage_HS_gzz = [98, 94, 99, 99, 97]
coverage_SS_gzz = [96, 94, 99, 99, 95]
d = [50, 100, 150, 200, 250];

coverage_HS_stan = [98, 95, 98.67, 100, 98.4];
```

```julia
using PyPlot
fig = figure(figsize=(9,3))
tight_layout()
subplot(121)
plot(d-1, coverage_HS_gzz, "o", label="GZZ")
plot(d+1, coverage_HS_stan, "o", label="HMC")
grid(alpha=0.5)
legend(frameon=true, ncol=2)
axis("tight")
title("Horseshoe prior", fontsize=11)
xlabel("Dimension", fontsize=10)
ylabel("Coverage", fontsize=10)
xticks(d);
subplot(122)
plot(d, coverage_SS_gzz, "o")
grid(alpha=0.5)
axis("tight")
title("Spike-and-slab prior", fontsize=11)
suptitle("Coverage of 90% posterior credible intervals", fontsize=13)
xlabel("Dimension", fontsize=10)
xticks(d)
subplots_adjust(top=0.84)
savefig("coverage.pdf", format="pdf", dpi=10000, bbox_inches="tight")
```

```julia
HS_results = load("data/HS_results.jld")
```

```julia
xi_samples_HS_gzz = HS_results["xi_samples"]
maxlag = 1000
acfs = []
for i in 1:length(xi_samples_HS_gzz) 
    d = size(xi_samples_HS_gzz[i],1)
    a = zeros(d, maxlag)
    for dim in 1:d 
        a[dim,:] = acf(xi_samples_HS_gzz[i][dim,:], maxlag)
    end
    push!(acfs,a)
end
```

```julia
for i in 1:length(xi_samples_HS_gzz) 
    print("Mean ESS: ", mean(size(xi_samples_HS_gzz[1],2)./abs.(sum(acfs[i],2))), "\n")
end
```
