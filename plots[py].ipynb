```python
import numpy as np, pystan as ps, numpy.random as npr, matplotlib.pyplot as plt, h5py
%matplotlib inline 
from time import time
from pylab import plot, show, legend
from stan_temp import *
```

```python
HS_results_stan = np.load("data/HS_results_stan.npy")
ci_HS_stan = HS_results_stan.item(0)['ci']
```

```python
k = 1
data = h5py.File("data/GZZ_data"+str(k+1)+".jld", "r")
ξ_true = data["xi_true"].value
data = h5py.File("data/HS_ci_gzz"+str(k+1)+".jld", "r")
ci_HS_gzz = data["ci"].value
dim = len(ξ_true)

fig = plt.subplots(figsize=(17,10))
plt.subplots_adjust(hspace=0.3)

ymin, ymax = -17, 21

plt.subplot(211)
for i in range(np.shape(ci_HS_stan[k])[0]) :
    plt.plot([i+1,i+1], ci_HS_stan[k][i], color="blue", lw=1.5)
plt.plot(np.arange(1,np.shape(ci_HS_stan[k])[0]+1), ξ_true, "ro", markersize=4)
plt.grid(True)
plt.title("Hamiltonian Monte Carlo", fontsize=14)
plt.xlabel(r"$\xi$ component", fontsize=12)
plt.ylabel("Credible interval", fontsize=12)
plt.ylim([ymin,ymax])

plt.subplot(212)
for i in range(np.shape(ci_HS_gzz)[1]) :
    plt.plot([i+1,i+1], ci_HS_gzz[:,i], color="blue", lw=1.5)
plt.plot(np.arange(1,np.shape(ci_HS_gzz)[1]+1), ξ_true, "ro", markersize=4)
plt.grid(True)
plt.title("Gibbs zig-zag sampler", fontsize=14)
plt.xlabel(r"$\xi$ component", fontsize=12)
plt.ylabel("Credible interval", fontsize=12)
plt.ylim([ymin,ymax])

plt.suptitle("90% credible intervals while using horeshoe prior when dimension = "+str(dim), fontsize=16, y=0.95)

plt.savefig("plots/CI_HS"+str(k+1)+".pdf", dpi=1000, bbox_inches="tight")
```
